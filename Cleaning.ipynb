#### First, the quotes

```python
df_all_states['final_words_as_str'] = df_all_states['final_words'].str.findall(r"(\".*?\")")
df_all_states.final_words_as_str
```

```python
df_all_states.final_words_as_str = df_all_states.final_words_as_str.astype(str).str.replace(']','')
df_all_states.final_words_as_str = df_all_states.final_words_as_str.astype(str).str.replace('[','')
df_all_states.final_words_as_str = df_all_states.final_words_as_str.astype(str).str.replace('\\','')

#df_all_states.final_words_as_str = df_all_states.final_words_as_str.str.replace(']','')
```

```python
df_all_states.final_words_as_str
```

```python
def replace_quotes_parens(row):
    line = row['final_words_as_str']
    replaced = re.sub("\'\"",'\"', line)
    replaced = re.sub("\"\'",'\"', replaced)
    return replaced
```

```python
df_all_states['final_words_cleaned'] = df_all_states.apply(replace_quotes_parens, axis=1)
```

```python
df_all_states
```

```python
# Instead of replacing all the stuff in the date we'll leave it as it is - for now.
# The only thing we'll do is  that we'll change the / to being a 
# and the ' - ' to being 'until
df_all_states.date_offense.replace({r"\s\-\s":" until ",
                                          r"\s\/\s":" & ",}, inplace=True, regex=True)
pd.set_option('max_colwidth', 400)

df_all_states
```

```python
# Now we need to:
# Ad a column with all WORDS of the final_words_cleaned lowercase and cleaned (no punctuation etc etc)
# And replace NaN by unknown
# And blank by declined
```

```python
# Now let's replace column with everything NaN 
# WE WANT TO DO IT IN FINA_WORDS_CLEANED !!!
# NONE are where inmate declined or his words were not reported as directly pronounced by him
df_all_states.final_words_cleaned.replace('nan', 'None', inplace=True)
# Unknown are we have no words at all
df_all_states.final_words_cleaned.replace('', 'Unknown', inplace=True)
df_all_states
```

```python
# Let's add column with everything cleaned - but will need to tokenize though if we want to do text analysis

def strip_all_stuff(row):
    line = row['final_words_cleaned']
    replaced = re.sub('[\"\.\,\!\?\-\*\:\;\(\)\_\'\+\=\&]',' ', line)
    replaced = replaced.lower()
    replaced = re.sub("\"\'",'\"', replaced)
    return replaced


df_all_states['final_words_str_tokenized'] = df_all_states.apply(strip_all_stuff, axis=1)
```

```python
df_all_states.head(50)
```

```python
# Now I want 3 additional columns: one that mentions 'innocent', another one that mentions 'love' another one that mentions 'god' another one that mentions 'racism'
```

```python
# Checking for love

def contains_love(row):
    line = row['final_words_str_tokenized'] 
    if ' love ' in line:
        return True
    else:
        return False
    
df_all_states['contains_love'] = df_all_states.apply(contains_love, axis=1)
df_all_states
```

```python
# This should avoid containing the things when the inmate talk about the other person's family

def contains_family(row):
    line = row['final_words_str_tokenized']
    words = ["my baby", "my baby", "my babies", "my kids", "my kid", "my son", "my sons", "my daughter", "my daughters", "auntie", "my family", "my home", "my children", "my child"]
    if any( x in line for x in words):
        return True
    else:
        return False

df_all_states['contains_family'] = df_all_states.apply(contains_family, axis=1)
```

```python
# Checking for innocence

def contains_innocent(row):
    line = row['final_words_str_tokenized']
    words = ["innocent", "innocence", "didn t do it", "didn t kill", "didn t commit", "innocence,"]
    if any( x in line for x in words):
        return True
    else:
        return False
    
df_all_states['contains_innocent'] = df_all_states.apply(contains_innocent, axis=1)
df_all_states
```

```python
# Checking for god

def contains_god(row):
    line = row['final_words_str_tokenized']
    words = [' god ', ' allah ', ' christ ', 'jesus', ' amen ']
    if any( x in line for x in words):
        return True
    else:
        return False
    
df_all_states['contains_god'] = df_all_states.apply(contains_god, axis=1)
df_all_states
```

```python
# Checking for remorse

def contains_remorse(row):
    line = row['final_words_str_tokenized']
    words = ['i regret', 'have regrets', 'have regrets', ' remorse ', 'm sorry', 'm so sorry', 'i apologize', 'i regret', 'i fully regret', 'have regrets', 'to apologize', 'have remorse', 'hearfelt remorse', 'much remorse', 'so much remorse', 'a lot of regrets', 'much regrets', 'many regrets', 'how sorry']
    if any( x in line for x in words):
        return True
    else:
        return False
    
df_all_states['contains_remorse'] = df_all_states.apply(contains_remorse, axis=1)
df_all_states
```

```python
# Checking for remorse

def contains_race(row):
    line = row['final_words_str_tokenized']
    words = [' racism ', ' racist ', ' racists ', ' discrimination ', ' discriminated ', ' black ', ' white ', ' african ', 'all white',' african american ']
    if any( x in line for x in words):
        forbidden_words = [' ass ', ' chris ']
        for each in forbidden_words:
            if each in line:
                return False
        return True
    else:
        return False
    
df_all_states['contains_race'] = df_all_states.apply(contains_race, axis=1)
df_all_states.tail(500)
```

```python
# RE TOKENIZING 
def tokenize(s):
    blob = TextBlob(s.lower())
    words = [token for token in blob.words if len(token)>1]
    return words

def print_sorted_vector(v):
    # this "lambda" thing is an anonymous function, google me to unluck bonus coding knowledge
    sorted_list = sorted(v.items(), key=lambda x: (x[1],x[0]), reverse=True) 
    sorted_list = sorted_list[:20]
    print('\n'.join([str(x) for x in sorted_list]))
```

```python
# tfidf = something

vectorizer = TfidfVectorizer(stop_words='english', tokenizer=tokenize)

# Use the vectorizor we just made. The name fit_transform will be clearer later when we use it for machine learning
matrix = vectorizer.fit_transform(analyze_african_americans["final_words_str_tokenized"])

# The easiest way to see what happenned is to make a dataframe
tfidf = pd.DataFrame(matrix.toarray(), columns=vectorizer.get_feature_names())
tfidf.head(50)
```

```python
docs = tfidf.iloc[:]
total = docs.sum(axis=0)
print_sorted_vector(total)
```
